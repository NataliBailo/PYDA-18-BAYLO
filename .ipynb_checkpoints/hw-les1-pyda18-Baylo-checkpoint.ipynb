```python
"""
Задание 2. Проверка високоссный год или нет.
Год високосный если он делится на 4 без остатка и с остатком делится на 100 или без остатка 
делится на 400. 

Для проверки использовались значения года из примера к заданию:
year = 2020
2020-високосный
year = 2019
2019-обычный
"""
year = 2024
if year%4 == 0 and year%100 != 0 or year%400 == 0:
    print (year, '- високосный год')
else:
    print (year, '- обычный год')
```

```python
"""
Задание 3. Программа просит последовательно ввеести день и месяц рождения, по которым определяет
и позвращает пользователю знак зодиака.

Для проверки использовались значения года из примера к заданию:
день: 30
месяц: Август
Ваш знак зодиака: Дева
и
день:29
месяц: Октябрь
Ваш знак зодиака: Скорпион

"""

day = int(input('Введите день:'))
month = str(input('Введите месяц (с большой буквы в именительном падеже):'))
#Овен: 21 марта — 20 апреля
if month == 'Март' and (
        day >= 21 and day <= 31) or month == 'Апрель' and (
        day >= 1 and day <= 20):
    print('Ваш знак зодиака: Овен')
#Телец: 21 апреля — 21 мая
elif month == 'Апрель' and (
        day >= 21 and day <= 30) or month == 'Май' and (
        day >= 1 and day <= 20):
    print('Ваш знак зодиака: Телец')
#Близнецы: 22 мая — 21 июня
elif month == 'Май'and (
        day >= 22 and day <= 31) or month == 'Июнь' and (
        day >= 1 and day <= 21):
    print ('Ваш знак зодиака: Близнецы')   
#Рак: 22 июня — 22 июля
elif month == 'Июнь'and (
        day >= 23 and day <= 30) or month == 'Июль' and (
        day >= 1 and day <= 22):
    print ('Ваш знак зодиака: Рак')
#Лев: 23 июля — 21 августа
elif month == 'Июль'and (
        day >= 23 and day <= 31) or month == 'Август' and (
        day >= 1 and day <= 21):
    print ('Ваш знак зодиака: Лев')
#Дева: 22 августа — 23 сентября
elif month == 'Август'and (
        day >= 22 and day <= 31) or month == 'Сентябрь' and (
        day >= 1 and day <= 23):
    print ('Ваш знак зодиака: Дева')
#Весы: 24 сентября — 23 октября
elif month == 'Сентябрь'and (
        day >= 24 and day <= 30) or month == 'Октябрь' and (
        day >= 1 and day <= 23):
    print ('Ваш знак зодиака: Весы')
#Скорпион: 24 октября — 22 ноября
elif month == 'Октябрь'and (
        day >= 24 and day <= 31) or month == 'Ноябрь' and (
        day >= 1 and day <= 22):
    print ('Ваш знак зодиака: Скорпион')
#Стрелец: 23 ноября — 22 декабря
elif month == 'Ноябрь'and (
        day >= 23 and day <= 30) or month == 'Декабрь' and (
        day >= 1 and day <= 22):
    print ('Ваш знак зодиака: Стрелец')
#Козерог: 23 декабря — 20 января
elif month == 'Декабрь'and (
        day >= 23 and day <= 31) or month == 'Январь' and (
        day >= 1 and day <= 20):
    print ('Ваш знак зодиака: Козерог')
#Водолей: 21 января — 19 февраля
elif month == 'Янваорь'and (
        day >= 21 and day <= 31) or month == 'Февраль' and (
        day >= 1 and day <= 19):
    print ('Ваш знак зодиака: Водолей')
#Рыбы: 20 февраля — 20 марта
elif month == 'Февраль'and (
        day >= 20 and day <= 29) or month == 'Март' and (
        day >= 1 and day <= 20):
    print ('Ваш знак зодиака: Рыбы')
else: print ('Введен неправильный формат даты и/или месяца')
```

```python
"""
Задание 4

Программа для подбора упаковок по размерам товара. 
Размеры (ширина, длина, высота) хранятся в переменных (в сантиметрах):

Правила подбора:
если каждое из трех измерений менее 15 сантиметров, то выведите на экран "Коробка №1";
если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то 
выводите "Коробка №2";
если длина товара больше 2 метров, то выводите "Упаковка для лыж";
во всех остальных случаях выводите "Стандартная коробка №3".

Для проверки использовались значения из примера к заданию:

width = 15
length = 50
height = 15
Стандартная коробка №3

width = 45
length = 205
height = 45
Упаковка для лыж
"""

width = int(input('Введите ширину (в см.):'))
length = int(input('Введите длину (в см.):'))
height = int(input('Введите высоту (в см.):'))
# "Коробка №1"
if width < 15 and length < 15 and height < 15:
    print ('Коробка №1')
# "Коробка №2"
elif height >= 15 and height < 50 and (
    width < 15 and length < 15) or length >= 15 and length < 50 and (
    width < 15 and height < 15) or width >= 15 and width < 50 and(
    length < 15 and height < 15):
    print ('Коробка №2')
# "Упаковка для лыж"
elif length > 200:
    print('Упаковка для лыж')
# "Стандартная коробка №3"    
else: print ('Стандартная коробка №3')
```

```python
"""
Задание 5 (необязательное)
Дана переменная, в которой хранится шестизначное число (номер проездного билета). 
Напишите программу, которая будет определять, является ли данный билет "счастливым".
Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.

Для проверки использовались значения из примера к заданию:

number = 123456
Результат:
Неасчастливый билет

number = 123321
Результат:
Счастливый билет
"""
#Тут я запрашиваю номер билета, а не присваиваю в качестве Дано
number = str(input('Введите шесть номер билета:'))
first_3_nunbers_sum = int(number[0]) + int(number[1]) + int(number[2])
second_3_nunbers_sum = int(number[3]) + int(number[4]) + int(number[5])
if first_3_nunbers_sum == second_3_nunbers_sum:
    print ('Счастливый')
else:
    print ('Обычный')
```

```python
"""Задание 6 (необязательное)
Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). 
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

если пользователь выбрал круг, запрашиваем его радиус,
если треугольник – длины трех его сторон;
если прямоугольник – длины двух его сторон.

Для проверки использовались значения из примера к заданию:

Круг
Введите радиус круга:10
Площадь круга: 314.16


Треугольник
Введите длину стороны A: 2
Введите длину стороны B: 2
Введите длину стороны C:3
Площадь треугольника: 1.98
"""
import math

figure = str(input('Введите тип фигуры (все буквы строчные):'))
if figure == 'круг':
    circle_radius = int(input('Введите радиус круга: '))
    print ('Площадь круга = ', math.pi*(circle_radius**2))

elif figure == 'треугольник':
    lenght_A = int(input('Введите длинну стороны A: '))
    lenght_B = int(input('Введите длинну стороны B: '))
    lenght_C = int(input('Введите длинну стороны C: '))
    #площадь треугольника по формуле Герона
    semi_perimeter = (lenght_A + lenght_B + lenght_C)/2
    print ('Площадь треугольника = ', math.sqrt(semi_perimeter*(semi_perimeter - lenght_A)*(semi_perimeter - lenght_B)*(semi_perimeter - lenght_C)))

elif figure == 'прямоугольник':
    lenght_A = int(input('Введите длинну стороны A: '))
    lenght_B = int(input('Введите длинну стороны B: '))
    print ('Площадь прямоугольника = ', lenght_A*lenght_B)
```
